# Modifications after Webex Meeting with Dow Team + Dr. Irfan Khan, May 19, 2020: Implicit Time-Stepping

On Tuesday, May 19, 2020, I met with members of our Dow team and Dow scientist Dr. Irfan Khan over Webex to discuss next steps for improving this model of bubble growth. This notebook implements and records the results of one those improvements: implicit time-stepping. Dr. Irfan Khan suggested using an implicit method having noticed problems with stability even at smaller time steps. This method implements the implicit Euler method.

The other improvement is the implementation of the Tolman correction to the interfacial tension, which is explored in `20200520_eps_pless_tolman.ipynb`.

```python
# adds custom libraries to path
import sys
sys.path.append('../libs/')

# import libraries
import numpy as np
import matplotlib.pyplot as plt

import scipy.optimize
import scipy.interpolate

# custom libraries
import bubble
import flow
import plot
import analytics as an

from importlib import reload
reload(bubble)
reload(an)

# lists data files required
# load data (interfacial tension, solubility, and diffusivity) for polyols? we don't have data for VORANOL 360...use 1k3f
polyol_data_file = '1k3f_30c.csv'
# load CO2 equation of state data for desired temperature [csv]
eos_co2_file = 'eos_co2_22-0C.csv'
# folder to save figures
save_folder = 'figures/'
save_figs = True

# system parameters
eta_o = 3.6 # viscosity of pure-polyol in outer stream, VORANOL 360, 'Polyol Properties.ppt'  [Pa.s]
p_in = 85E5 # estimate of pressure at capillary in let [Pa]
L = 0.099 # length of observation capillary, p. 154 of lab notebook #0005 [m]
p_atm = 101.3E3 # atmospheric pressure, assumed to be outlet pressure [Pa]
Q_o = 2000/60E9 # outer stream flow rate, input to ISCO 260 D [m^3/s]
Q_i = 180/60E9 # inner stream flow rate, input to ISCO 100 DM [m^3/s]
R_cap = 254E-6 # radius of inner walls of observation capillary, PFA 0.002" ID [m]
R_nuc = 4E-9 # initial bubble radius, estimated as a bit larger than prediction in slide 13 of 20200305_aps_march_meeting [m]
p_s = 72E5 # saturation pressure of CO2 in VORANOL 360, p. 154 of lab notebook #0005 [Pa]

# numerical parameters
dt = 1E-7 # time step for numerical model [s]

# CONVERSIONS
s_2_ms = 1000
s_2_us = 1E6
kg_2_g = 1E3
kg_2_ng = 1E12
kgm3_2_gmL = 1E-3
m_2_mm = 1000
m_2_um = 1E6
m_2_nm = 1E9
Pa_2_bar = 1E-5
Pa_2_MPa = 1E-6
Nm_2_mNm = 1000
```

### Solve for Flow Profile

Solve Stokes flow and constraints from conservation of volume for inner and outer streams.

```python
# calculates inner stream velocity based on flow rates, viscosities, and observation capillary inner diameter
eta_i0 = 0.8
R_i0 = 50E-6
v0 = 0.5
x0 = [eta_i0, R_i0, v0] # initial guesses for p_in [Pa], v [m/s], R_i [m]
ordering = [0, 3, 4, 5, 6, 7, 8, 1, 9, 2] # indices of vars in alphabetical order from list vars + args
args = (eta_o, L, p_atm, p_in, Q_i, Q_o, R_cap, ordering)
soln = scipy.optimize.root(flow.sheath_eqns_input, x0, args=(args,))
eta_i, R_i, v = soln.x

print('Inner stream viscosity eta_i = {0:.2f} Pa.s'.format(eta_i))
print('Inner stream velocity v = {0:.1f} mm/s'.format(v*m_2_mm))
print('Radius of inner stream R_i = {0:.1f} um.'.format(R_i*m_2_um))
```

### Test Bubble Growth Model: Explicit

```python
reload(bubble)
t_nuc = 0.068
dt = 1E-3
# computes bubble growth trajectory using modified Epstein-Plesset model
t, m, D, p, p_bubble, if_tension, c_s, R, rho_co2 = bubble.grow(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                                            polyol_data_file, eos_co2_file, 
                                                                            adaptive_dt=True, implicit=False)
# records results from explicit Euler method
t_expl = (np.array(t)-t_nuc)*s_2_ms
R_expl = np.array(R)*m_2_um
```

Below, we plot the results to check if they make physical sense.

```python
props_list = (R, m, p, p_bubble, rho_co2, if_tension)
ax = plot.all_props(t, t_nuc, props_list, x_log=True, title='Explicit')
```

This result match the previous results, *e.g.*, in `20200518_eps_pless_jak_mods.ipynb`.

### Compare to Implicit

Now we compare the results of explicit Euler method to the implicit Euler method for time stepping.

```python
reload(bubble)
# performs explicit Euler method
t, m, D, p, p_bubble, if_tension, c_s, R, rho_co2 = bubble.grow(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                                            polyol_data_file, eos_co2_file, 
                                                                            adaptive_dt=True, implicit=True)
# records results from explicit Euler method
t_impl = (np.array(t)-t_nuc)*s_2_ms
R_impl = np.array(R)*m_2_um
```

Plots results.

```python
# plots difference
R_expl_interp = np.interp(t_impl, t_expl, R_expl)
diff = np.abs(R_impl - R_expl_interp)
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t_impl, R_impl, label='implicit')
ax.plot(t_expl, R_expl, label='explicit')
ax.plot(t_impl, diff, label='diff')
ax.plot(t_impl, diff/R_impl, '--', label='frac diff (nondim)')
ax.set_xlabel(r'$t-t_{nuc}$ [ms]', fontsize=16)
ax.set_ylabel(r'$R(t)$ [$\mu$m]', fontsize=16)
ax.set_title('Implicit vs. Explicit Euler'.format(dt*s_2_us), fontsize=18)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim([1E-6, 1E3])
ax.tick_params(axis='both', labelsize=14)
plt.legend(fontsize=14)
```

Now that I use the result from the explicit time step for the initial guess of the implicit time step, the differences between them are much smaller.

```python
reload(an)
# explicit time-stepping for different tolerances on R
growth_model = bubble.grow
tol_R_list = [1E-4, 3E-4, 1E-3, 3E-3, 0.01, 0.03, 0.1]
t_nuc = 0.068
adaptive_dt = True
implicit = False

expl = an.tol_R_convergence(growth_model, tol_R_list, t_nuc, p_s, R_nuc, p_atm, L, 
                          p_in, v, polyol_data_file, eos_co2_file, implicit)
```

We extract the radius, time and pressure from the output.

```python
# extracts results
t_list_expl = expl[0]
R_list_expl = expl[-2]
p_list_expl = expl[3]
p_bubble_list_expl = expl[4]
```

Finally, we plot to compare the trajectory of bubble growth for different time steps.

```python
label_list = ['tol_R = {0:f}'.format(tol_R) for tol_R in tol_R_list]
ax = plot.series(t_list_expl, R_list_expl, label_list, t_nuc, s_2_ms, m_2_um, r'$t - t_{nuc}$ [ms]',
           r'$R$ [$\mu$m]', 'R Tolerance Convergence, Explicit', x_log=True, y_log=True)
```

The trajectory converges within the width of the plot line for $tol_R \le 0.03$, with the largest (fractional) discrepancies occurring at the beginning.

I will zoom in on the initial growth for the three best estimates to see the convergence more clearly.

```python
reload(plot)

ax = plot.series(t_list_expl[:3], R_list_expl[:3], label_list[:3], t_nuc, s_2_ms, m_2_um, r'$t - t_{nuc}$ [ms]',
           r'$R$ [$\mu$m]', 'R Tolerance Convergence, Explicit', x_log=True, y_log=True)
ax.set_xlim([1E-5, 1E-2])
ax.set_ylim([3E-3, 1E-1])
```

Even after zooming in on the three closest estimates, it is difficult to discern how much the convergence is improving quantitatively with the lines overlapping, so I will plot the differences below.

```python
reload(plot)
diff_label_list = label_list
diff_label_list[0] = 'ref, tol_R = {0:f}'.format(tol_R_list[0])
ax = plot.diff(t_list_expl, R_list_expl, diff_label_list, t_nuc, s_2_ms, m_2_um, r'$t - t_{nuc}$ [ms]',
           r'$R$ [$\mu$m]', 'Error for R Tolerance Convergence, Explicit', x_log=True, y_log=True)
```

The difference from the reference (the smallest tolerance on $R$, $tol_R = 0.0001$) only decreases with decreasing tolerance on $R$ after the initial nucleation. To have a better idea of the relative errors throughout the trajectory, I will plot the fractional error below.

```python
reload(plot)
ax = plot.diff(t_list_expl, R_list_expl, diff_label_list[1:], t_nuc, s_2_ms, 1, r'$t - t_{nuc}$ [ms]',
           r'$\Delta R/R_{ref}$ [nondim]', 'Error for R Tolerance Convergence, Explicit', x_log=True, y_log=True, frac=True)
ax.set_ylim([1E-5, 1])
```

Besides all having relative errors around 0.1% at the beginning, the fractional errors are roughly the size of the tolerances on $R$, as desired.

#### Implicit

Now we repeat the above steps for the implicit Euler method.

```python
reload(an)
# explicit time-stepping for different tolerances on R
growth_model = bubble.grow
tol_R_list = [1E-4, 3E-4, 1E-3, 3E-3, 0.01, 0.03, 0.1]
t_nuc = 0.068
adaptive_dt = True
implicit = True

impl = an.tol_R_convergence(growth_model, tol_R_list, t_nuc, p_s, R_nuc, p_atm, L, 
                          p_in, v, polyol_data_file, eos_co2_file, implicit)
```

We extract the radius, time and pressure from the output.

```python
# extracts results
t_list_impl = impl[0]
R_list_impl = impl[-2]
p_list_impl = impl[3]
p_bubble_list_impl = impl[4]
```

Finally, we plot to compare the trajectory of bubble growth for different time steps.

```python
label_list = ['tol_R = {0:f}'.format(tol_R) for tol_R in tol_R_list]
ax = plot.series(t_list_impl, R_list_impl, label_list, t_nuc, s_2_ms, m_2_um, r'$t - t_{nuc}$ [ms]',
           r'$R$ [$\mu$m]', 'R Tolerance Convergence, Implicit', x_log=True, y_log=True)
```

The results look identical to those for explicit time-stepping. Let's zoom into the beginning for a closer look.

```python
ax = plot.series(t_list_impl[:3], R_list_impl[:3], label_list[:3], t_nuc, s_2_ms, m_2_um, r'$t - t_{nuc}$ [ms]',
           r'$R$ [$\mu$m]', 'R Tolerance Convergence, Implicit', x_log=True, y_log=True)
```

Even zoomed-in, I still cannot distinguish a difference between the predictions of the bubble growth for low tolerance on $R$. Let's plot the differences now.

```python
ax = plot.diff(t_list_impl, R_list_impl, diff_label_list[1:], t_nuc, s_2_ms, 1, r'$t - t_{nuc}$ [ms]',
           r'$\Delta R/R_{ref}$ [nondim]', 'Error for R Tolerance Convergence, Implicit', x_log=True, y_log=True, frac=True)
ax.set_ylim([1E-5, 1])
```

Once again, the errors look similar to those for the explicit time-stepping: around 0.1% at the beginning for all tolerances, followed by an error approximately equal to the tolerance on $R$. The errors get much lower at some points, however, and for some tolerances the error decreases at the very end, opposite the case for the explicit time-stepping.

To compare implicit vs. explicit time-stepping more concretely, we will plot the difference below.

#### Implicit vs. Explicit

```python
i_ref = 0
t_list_comp = [t_list_expl[i_ref], t_list_impl[i_ref]]
R_list_comp = [R_list_expl[i_ref], R_list_impl[i_ref]]
label_list_comp = ['expl - impl']
ax = plot.diff(t_list_comp, R_list_comp, label_list_comp, t_nuc, s_2_ms, 1, r'$t - t_{nuc}$ [ms]',
           r'$\Delta R/R_{ref}$ [nondim]', 'Explicit vs. Implicit, tol_R = {0:f}'.format(tol_R_list[i_ref]), 
               x_log=True, y_log=True, frac=True)
ax.set_ylim([1E-7, 1])
```

The discrepancy between implicit and explicit time-stepping methods is roughly the tolerance on $R$, in this case, 0.01%. The only point at which the discrepancy is large is at the jump in the density, which is probably because the jump is predicted to occur one time step earlier for the implicit time-stepping method.

**Therefore, it appears that for this model, the difference between implicit and explicit time-stepping is not significant.** Since the explicit method is faster (due to the use of the explicit method to provide the initial guesses for the implicit method), I will probably use this in future studies with this model. Nevertheless, it is possible that in some cases implicit time-stepping will be required, in which case the model is ready.

To drive home that the implicit and explicit models make similar predictions, I plot the results from both time-stepping methods below.

```python
ax = plot.series(t_list_comp, R_list_comp, ['expl', 'impl'], t_nuc, s_2_ms, m_2_um, r'$t - t_{nuc}$ [ms]',
           r'$R$ [$\mu$m]', 'Explicit vs. Implicit, R Tolerance = {0:f}'.format(tol_R_list[i_ref]), x_log=True, y_log=True)
```

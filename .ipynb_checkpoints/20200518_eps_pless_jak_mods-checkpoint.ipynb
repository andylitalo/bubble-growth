# Modifications after Meeting with Julie on May 13, 2020

This notebook implements modifications suggested during my meeting with Julie on May 13, 2020. These modifications include

1. extrapolating the interfacial tension to high pressures (I did this with a linear extrapolation, although Julie suggested using DFT predictions)
2. uses pressure inside bubble to estimate interfacial tension. Eventually will have DFT look-up table.???
3. Also, doesn't update diffusivity, but keeps the saturation value based on the assumption that the concentration of CO2 in the bulk doesn't change substantially (which seems consistent with our observation that a bubble  removes only a layer of < 2 um of CO2 from its surroundings).
4. The code is also faster because I do not load the data from the .csv files every time I want to make an interpolation, but instead load them once at the beginning and just interpolate with the resulting numpy arrays.
5. Additionally, uses t_i to calculate dm/dt_i in the explicit Euler method, instead of t_{i+1}. This requires integrating the 1/np.sqrt(t) in the first time step.

These changes are implemented in `eps_pless_p_if_4()` in the `bubble.py` library.

Additionally,

6. Deriving the flow properties using an estimate for the pressure at the inlet to the observation capillary, $p_{in}$, instead of estimating the inner stream's viscosity $\eta_i$.

This is implemented by `flow.sheath_eqns_input()`.

```python
# adds custom libraries to path
import sys
sys.path.append('../libs/')
sys.path.append('src/')

# import libraries
import numpy as np
import matplotlib.pyplot as plt

import scipy.optimize
import scipy.interpolate

# custom libraries
import bubblelegacy as bubble
import flow

from importlib import reload
reload(bubble)

# lists data files required
# load data (interfacial tension, solubility, and diffusivity) for polyols? we don't have data for VORANOL 360...use 1k3f
polyol_data_file = 'input/1k3f_30c.csv'
# load CO2 equation of state data for desired temperature [csv]
eos_co2_file = 'input/eos_co2_22-0C.csv'
# folder to save figures
save_folder = 'figs/'
save_figs = True

# system parameters
eta_o = 3.6 # viscosity of pure-polyol in outer stream, VORANOL 360, 'Polyol Properties.ppt'  [Pa.s]
p_in = 85E5 # estimate of pressure at capillary in let [Pa]
L = 0.099 # length of observation capillary, p. 154 of lab notebook #0005 [m]
p_atm = 101.3E3 # atmospheric pressure, assumed to be outlet pressure [Pa]
Q_o = 2000/60E9 # outer stream flow rate, input to ISCO 260 D [m^3/s]
Q_i = 180/60E9 # inner stream flow rate, input to ISCO 100 DM [m^3/s]
R_cap = 254E-6 # radius of inner walls of observation capillary, PFA 0.002" ID [m]
R_nuc = 4E-9 # initial bubble radius, estimated as a bit larger than prediction in slide 13 of 20200305_aps_march_meeting [m]
p_s = 72E5 # saturation pressure of CO2 in VORANOL 360, p. 154 of lab notebook #0005 [Pa]

# numerical parameters
dt = 3E-6 # time step for numerical model [s]
t_nuc = 0 # initial time of bubble nucleation after entering observation capillary [s]

# CONVERSIONS
s_2_ms = 1000
kg_2_g = 1E3
kg_2_ng = 1E12
kgm3_2_gmL = 1E-3
m_2_mm = 1000
m_2_um = 1E6
m_2_nm = 1E9
Pa_2_bar = 1E-5
Pa_2_MPa = 1E-6
Nm_2_mNm = 1000
```

We must next derive the following input parameters.

```python
# calculates inner stream velocity based on flow rates, viscosities, and observation capillary inner diameter
eta_i0 = 0.8
R_i0 = 50E-6
v0 = 0.5
x0 = [eta_i0, R_i0, v0] # initial guesses for p_in [Pa], v [m/s], R_i [m]
ordering = [0, 3, 4, 5, 6, 7, 8, 1, 9, 2] # indices of vars in alphabetical order from list vars + args
args = (eta_o, L, p_atm, p_in, Q_i, Q_o, R_cap, ordering)
soln = scipy.optimize.root(flow.sheath_eqns_input, x0, args=(args,))
eta_i, R_i, v = soln.x

print('Inner stream viscosity eta_i = {0:.2f} Pa.s'.format(eta_i))
print('Inner stream velocity v = {0:.1f} mm/s'.format(v*m_2_mm))
print('Radius of inner stream R_i = {0:.1f} um.'.format(R_i*m_2_um))
```

The numbers for pressure, velocity, and inner stream radius make sense.

In this example, a CO2 bubble is artificially nucleated at the entrance of the observation capillary, where the pressure is about 80 bar, despite the saturation pressure being around 72 bar (below we identify a more reasonable nucleation pressure around 58 bar). The pressure inside the bubble is only noticeably higher than the pressure outside the bubble until the bubble starts to grow. This gives me confidence that the density inside the bubble will not be more vapor-like by the time it reaches 50 nm, which should allow us to detect it with SAXS. The initial nuclei, however, have liquid-like densities, and thus will be difficult to detect, especially given that they will occupy less total volume as well (so the intensity, which is proportional to the square of the density difference times the occupied volume of the bubbles, $I \propto (\Delta \rho V)^2$, will be lowered by both factors during the initial nucleation and growth).

The bubble growth skyrockets at the end as, fractionally, the pressure drop accelerates, as seen by the downturn of the pressure at the end. The final bubble is about 130 $\mu$m in radius, or 260 $\mu$m in diameter, roughly the size observed in polyurethane foams produced under similar conditions.


We now want to make sure that the results converge as the time step is decreased.

```python
dt_list = [1E-6, 0.000003, 0.00001, 0.00003, 0.0001, 0.0003, 0.001, 0.003, 0.01]
t_list = []
m_list = []
D_list = []
p_list = []
p_bubble_list = []
if_tension_list = []
c_s_list = []
R_list = []
rho_co2_list = []

t_nuc = 0.0676

# solves bubble growth for different time steps
for dt in dt_list:
    t, m, D, p, p_bubble, if_tension, c_s, R, rho_co2 = bubble.eps_pless_p_if_4(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                                  polyol_data_file, eos_co2_file, adaptive_dt=True)
    t_list += [t]
    m_list += [m]
    D_list += [D]
    p_list += [p]
    p_bubble_list += [p_bubble]
    if_tension_list += [if_tension]
    c_s_list += [c_s]
    R_list += [R]
    rho_co2_list += [rho_co2]
    print('completed iteration for dt = {0:f}'.format(dt))
```

```python
# plots R(t)
fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(len(dt_list)):
    dt = dt_list[i]
    t = t_list[i]
    R = R_list[i]
    ax.plot(np.array(t)*s_2_ms, np.array(R)*m_2_um, label='dt = {0:.3f} ms'.format(dt*s_2_ms))
    
ax.set_xlim([67, 75])
ax.set_yscale('log')
ax.set_xlabel('time [ms]', fontsize=18)
ax.set_ylabel(r'$R(t)$ [$\mu$m]', fontsize=18)
ax.set_title(r'$R(t)$ of CO2 Bubble in V360 for Different Time Steps', fontsize=20)
plt.legend()
```

The results do not converge near the nucleation time because of the almost infinite growth at the beginning. However, the results converge for $dt < 0.1$ ms for $R > 1~\mu$m. For this reason, I will change the time-stepping from 0.03 ms to 1 ms after that point.

```python
dt = 3E-6

# plots comparison between adaptive and fixed time-stepping
fig = plt.figure()
ax = fig.add_subplot(111)

# computes result with fixed time-stepping
t, m, D, p, p_bubble, if_tension, c_s, R, rho_co2 = bubble.eps_pless_p_if_4(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                              polyol_data_file, eos_co2_file, adaptive_dt=False)
ax.plot(np.array(t)*s_2_ms, np.array(R)*m_2_um, label='dt = {0:.3f} ms'.format(dt*s_2_ms))

# computes result with adaptive time-stepping (increases time step after a threshold)
t_adpt, m_adpt, D_adpt, p_adpt, p_bubble_adpt, if_tension_adpt, c_s_adpt, R_adpt, rho_co2_adpt = \
                                                    bubble.eps_pless_p_if_4(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                              polyol_data_file, eos_co2_file, adaptive_dt=True)
ax.plot(np.array(t_adpt)*s_2_ms, np.array(R_adpt)*m_2_um, label='dt = {0:.3f} ms, adpt'.format(dt_adpt*s_2_ms))

# plots difference
R_adpt_interp = np.interp(np.array(t), np.array(t_adpt), np.array(R_adpt))
ax.plot(np.array(t)*s_2_ms, np.abs(np.array(R)-R_adpt_interp)*m_2_um, label='diff')
    
# formats plot
ax.set_xlim([67, 90])
ax.set_yscale('log')
ax.set_xlabel('time [ms]', fontsize=18)
ax.set_ylabel(r'$R(t)$ [$\mu$m]', fontsize=18)
ax.set_title(r'$R(t)$ of CO2 Bubble in V360, Compare Adaptive $dt$', fontsize=20)
plt.legend()
```

Looks like I switched to the longer time step too soon, resulting in a 1 $\mu$m error right away. If I wait a little longer I should be able to reduce this error substantially.

Changed so that the step size gradually increases after reaching the threshold value of the radius and the error is now below 10 nm, which is acceptable.

**Use a step size of 0.03 ms with adaptive stepping.**

### Interfacial Tension Model

Given the small impact of the interfacial tension, I hypothesize that the model does not affect the result substantial. To test this hypothesis, I will compare the results for a linear fit to the interfacial tension data (underestimates at high pressures) with results for a flat function for the interfacial tension above the experimental data at 10 mN/m (overestimates at high pressures).

```python
reload(bubble)
# sets base level time step (increases with adaptive time-stepping) [s]
dt = 3E-6

# plots comparison between two models of interfacial tension (upper and lower bound)
fig = plt.figure()
ax = fig.add_subplot(111)

# computes result with lower bound on interfacial tension
t, m, D, p, p_bubble, if_tension, c_s, R, rho_co2 = bubble.eps_pless_p_if_4(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                              polyol_data_file, eos_co2_file, adaptive_dt=True,
                                                                           if_tension_model='lin')
ax.plot(np.array(t)*s_2_ms, np.array(R)*m_2_um, label='dt = {0:.3f} ms (linear)'.format(dt*s_2_ms))

# computes result with lower bound on interfacial tension
t2, m, D, p, p_bubble, if_tension, c_s, R2, rho_co2 = bubble.eps_pless_p_if_4(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                              polyol_data_file, eos_co2_file, adaptive_dt=True,
                                                                           if_tension_model='ceil')
ax.plot(np.array(t2)*s_2_ms, np.array(R2)*m_2_um, label='dt = {0:.3f} ms (flat)'.format(dt*s_2_ms))

# plots difference
R2_interp = np.interp(np.array(t), np.array(t2), np.array(R2))
ax.plot(np.array(t)*s_2_ms, np.abs(np.array(R)-R2_interp)*m_2_um, label='diff')
    
# formats plot
ax.set_xlim([67, 90])
ax.set_yscale('log')
ax.set_xlabel('time [ms]', fontsize=18)
ax.set_ylabel(r'$R(t)$ [$\mu$m]', fontsize=18)
ax.set_title(r'$R(t)$ of CO2 Bubble in V360, Compare Interfacial Tension Models', fontsize=20)
plt.legend()
```

## Tracing Back Bubble Growth

Ultimately, the goal of this bubble growth model is to estimate when a bubble nucleated based on the growth we can observe under the microscope. We can make this estimate using a modification of the shooting method, where we vary the initial time of bubble nucleation until the resulting trajectory matches a known data point for the radius at a given distance along the channel.

```python
reload(bubble)

# known measurement of bubble radius from Slide 13 of 20200305_aps_march_meeting.ppt
R_bubble = 12.5E-6 # bubble radius [m]
x_bubble = 48E-3 # distance traveled along channel [m]
t_bubble = x_bubble / v # time bubble had traveled down channel [s]
dt = 3E-5 # sets the time step to the optimal value found above--0.1 ms is reasonable after initial growth

# bounds on nucleation time
t_nuc_lo = 0.0 # [s]
t_nuc_hi = 0.12 # [s]

# tolerance on radius of bubble measurement [fractional], estimated by uncertainty in bubble measurement
tol_R = 0.01

fig = plt.figure()
ax = fig.add_subplot(111)

# uses modified shooting method to estimate the nucleation time
t_nuc = bubble.fit_growth_to_pt(t_bubble, R_bubble, t_nuc_lo, t_nuc_hi, dt, p_s, R_nuc, p_atm, L, p_in, v,
                     polyol_data_file, eos_co2_file, tol_R=tol_R, ax=ax, eps_pless_fn=bubble.eps_pless_p_if_4)
```

Under the new model, the bubble grows too quickly to match the fit point. Perhaps the model still requires modification, or perhaps the bubbles that were observed had nucleated before entering the channel.

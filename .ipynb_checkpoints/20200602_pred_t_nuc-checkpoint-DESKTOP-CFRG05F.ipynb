# Predicting the Nucleation Time by Tracing Back Bubble Growth

In this notebook, we explore how to estimate the nucleation time by tracing back bubble growth with a model developed in previous notebooks from experimental measurements of bubble sizes. We will also look at the sensitivity of the predicted nucleation to different parameters of the model.

**NOTE: I tried to reverse the `bubble.grow()` method from the finish to the nucleation by setting `dt < 0`, but realized that the method requires `t_nuc` to calculate `dmdt` because the time in the Epstein-Plesset model is measured from nucleation (whereas in these models I measure time from entering the microfluidic channel). For this reason, I must use the shooting method to estimate the nucleation time instead, despite its inefficiencies.**

```python
%%time

# import libraries
import numpy as np
import matplotlib.pyplot as plt

import scipy.optimize
import scipy.interpolate

# custom libraries
import bubble
import plot
import analytics as an

from importlib import reload
reload(bubble)
reload(an)

# lists data files required
# load data (interfacial tension, solubility, and diffusivity) for polyols? we don't have data for VORANOL 360...use 1k3f
polyol_data_file = '1k3f_22c.csv' # from DFT and PC-SAFT estimates
# load CO2 equation of state data for desired temperature [csv]
eos_co2_file = 'eos_co2_22-0C.csv'
# folder to save figures
save_folder = 'figures/'
save_figs = True

# system parameters
eta_o = 3.6 # viscosity of pure-polyol in outer stream, VORANOL 360, 'Polyol Properties.ppt'  [Pa.s]
p_in = 85E5 # estimate of pressure at capillary in let [Pa]
L = 0.099 # length of observation capillary, p. 154 of lab notebook #0005 [m]
p_atm = 101.3E3 # atmospheric pressure, assumed to be outlet pressure [Pa]
Q_o = 2000/60E9 # outer stream flow rate, input to ISCO 260 D [m^3/s]
Q_i = 180/60E9 # inner stream flow rate, input to ISCO 100 DM [m^3/s]
R_cap = 254E-6 # radius of inner walls of observation capillary, PFA 0.002" ID [m]
R_nuc = 4E-9 # initial bubble radius, estimated as a bit larger than prediction in slide 13 of 20200305_aps_march_meeting [m]
p_s = 72E5 # saturation pressure of CO2 in VORANOL 360, p. 154 of lab notebook #0005 [Pa]

# numerical parameters
dt = 1E-7 # time step for numerical model [s]
D = 1E-9 # [m^2/s]

# CONVERSIONS
s_2_ms = 1000
s_2_us = 1E6
kg_2_g = 1E3
kg_2_ng = 1E12
kgm3_2_gmL = 1E-3
m_2_mm = 1000
m_2_um = 1E6
m_2_nm = 1E9
Pa_2_bar = 1E-5
Pa_2_MPa = 1E-6
Nm_2_mNm = 1000
m2s_2_cm2s = 1E4
```

### Solve for Flow Profile

Solve Stokes flow and constraints from conservation of volume for inner and outer streams.

```python
%%time

# calculates inner stream velocity based on flow rates, viscosities, and observation capillary inner diameter
eta_i0 = 0.8
R_i0 = 50E-6
v0 = 0.5
x0 = [eta_i0, R_i0, v0] # initial guesses for p_in [Pa], v [m/s], R_i [m]
ordering = [0, 3, 4, 5, 6, 7, 8, 1, 9, 2] # indices of vars in alphabetical order from list vars + args
args = (eta_o, L, p_atm, p_in, Q_i, Q_o, R_cap, ordering)
soln = scipy.optimize.root(bubble.flow_eqns_input, x0, args=(args,))
eta_i, R_i, v = soln.x

print('Inner stream viscosity eta_i = {0:.2f} Pa.s'.format(eta_i))
print('Inner stream velocity v = {0:.1f} mm/s'.format(v*m_2_mm))
print('Radius of inner stream R_i = {0:.1f} um.'.format(R_i*m_2_um))
```

### Fitting Bubble Growth Trajectory to Point

I will use the shooting method to estimate a nucleation time `t_nuc` that results in a bubble growth trajectory that fits measurements of bubble size. 

```python
%%time

reload(bubble)
# known measurement of bubble radius from Slide 13 of 20200305_aps_march_meeting.ppt
R_bubble = 12.5E-6 # bubble radius [m]
x_bubble = 48E-3 # distance traveled along channel [m]
t_bubble = x_bubble / v # time bubble had traveled down channel [s]
dt = 1E-10 # sets the time step to the optimal value found above

# bounds on nucleation time
t_nuc_lo = bubble.calc_t_s(p_in, p_atm, p_s, v, L) # [s]
t_nuc_hi = 0.12 # [s]

# tolerance on radius of bubble measurement [fractional], estimated by uncertainty in bubble measurement
sigma_R = 0.01

fig = plt.figure()
ax = fig.add_subplot(111)

# computes bubble growth trajectory using modified Epstein-Plesset model
growth_fn = bubble.grow
adaptive_dt = True
if_tension_model = 'ceil'
implicit = False
d_tolman = 5E-9
tol_R = 1E-3
alpha = 0.5
args = [dt, p_s, R_nuc, p_atm, L, p_in, v, polyol_data_file, eos_co2_file, adaptive_dt, if_tension_model, 
        implicit, d_tolman, tol_R, alpha, D]
i_t_nuc = 1

# uses modified shooting method to estimate the nucleation time
t_nuc, results = an.fit_growth_to_pt(t_bubble, R_bubble, t_nuc_lo, t_nuc_hi, growth_fn, args,
                     i_t_nuc, sigma_R=sigma_R, ax=ax)

t, m, D, p, p_bubble, if_tension, c_s, R, rho_co2 = results
```

```python
%%time

# calculates the coordinates of another bubble for validation (@ 40 mm) [s]
t_bubble_2 = 0.04/v
# gives approximate bubble radius
R_bubble_2 = 10E-6

props_list = (R, m, p, p_bubble, rho_co2, if_tension)
ax = plot.all_props(t, t_nuc, props_list, title='Fit to Bubble Measurement')

# adds points of bubbles measured
ax.plot((t_bubble-t_nuc)*s_2_ms, R_bubble*m_2_um, color='k', marker='o', ms=6, label='fit pt')
ax.plot((t_bubble_2-t_nuc)*s_2_ms, R_bubble_2*m_2_um, color='m', marker='o', ms=6, label='validation pt')
ax.set_yscale('log')
ax.set_xlabel('time [ms]', fontsize=18)
ax.set_title('Bubble of CO2 Along Channel in V360', fontsize=20)
ax.set_ylim([0.001,1000])
# plt.legend()

# if save_figs:
#     plt.savefig(save_folder + 'co2_bubble_psat_72bar_V360.pdf', bbox_inches='tight')

# print out values of model at the measured points
print('Model prediction at fit point is R = {0:f} um.'.format(np.interp(t_bubble, t, np.array(R)*m_2_um)))
print('Model prediction at addl point is R = {0:f} um.'.format(np.interp(t_bubble_2, t, np.array(R)*m_2_um)))
```

According to this model, the two bubbles recorded could not have nucleated at the same location. The model appears to predict that the bubble used for fitting the data grew to the observed size within 1 ms of nucleation. I know that this is not true because the bubbles are in the frame for more than 1 ms but do not appear to change size. See `videos//20191212_v360_co2_72bar//v360_co2_1505fps_016us_2000o_180i_48mm_16.mp4` on the group server.

To get a clearer picture of what the model might predict, I will look at the predictions within the range of uncertainty for the diffusivity and for the size of the bubble.

```python
%%time

reload(bubble)
reload(an)

R_bubble_list = [11E-6, 12.5E-6, 14E-6] # range of bubble sizes [m]
D_list = [1E-9, 1.5E-9, 2E-9] # range of diffusivities [m^2/s]

t_nuc_list = []
t_list = []
m_list = []
p_list = []
p_bubble_list = []
if_tension_list = []
c_s_list = []
R_list = []
rho_co2_list = []

conditions_list = []

ctr = 0
for R_bubble in R_bubble_list:
    for D in D_list:
        args = [dt, p_s, R_nuc, p_atm, L, p_in, v, polyol_data_file, eos_co2_file, adaptive_dt, if_tension_model, 
                implicit, d_tolman, tol_R, alpha, D]

        # uses modified shooting method to estimate the nucleation time
        t_nuc, results = an.fit_growth_to_pt(t_bubble, R_bubble, t_nuc_lo, t_nuc_hi, growth_fn, args,
                             i_t_nuc, sigma_R=sigma_R)
        
        t_nuc_list += [t_nuc]

        t, m, _, p, p_bubble, if_tension, c_s, R, rho_co2 = results
        
        t_list += [t]
        m_list += [m]
        p_list += [p]
        p_bubble_list += [p_bubble]
        if_tension_list += [if_tension]
        c_s_list += [c_s]
        R_list += [R]
        rho_co2_list += [rho_co2]
        
        conditions_list += [(R_bubble, D)]
        
        ctr += 1
        print('Finished condition {0:d} of {1:d}.'.format(ctr, len(R_bubble_list)*len(D_list)))
```

Now plot the results at the lower and upper bounds for the diffusivity and the bubble size to fit, as well as at the center points.

```python
%%time

# calculates the coordinates of another bubble for validation (@ 40 mm) [s]
t_bubble_2 = 0.04/v
# gives approximate bubble radius
R_bubble_2 = 10E-6

for i, conditions in enumerate(conditions_list):
    R_bubble, D = conditions
    t_nuc = t_nuc_list[i]
    p_nuc = p_list[i][0] # nucleation pressure [Pa]
    props_list = (R_list[i], m_list[i], p_list[i], p_bubble_list[i], rho_co2_list[i], if_tension_list[i])
    ax = plot.all_props(t_list[i], t_nuc, props_list)

    # adds points of bubbles measured
    ax.plot((t_bubble-t_nuc)*s_2_ms, R_bubble*m_2_um, color='k', marker='o', ms=6, label='fit pt')
    ax.plot((t_bubble_2-t_nuc)*s_2_ms, R_bubble_2*m_2_um, color='m', marker='o', ms=6, label='validation pt')
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlabel('time [ms]', fontsize=18)
    ax.set_title('D = {0:g} cm^2/s, R = {1:.1f} um, t0 = {2:.1f} ms, p0 = {3:.1f} MPa'\
                 .format(D*m2s_2_cm2s, R_bubble*m_2_um, t_nuc*s_2_ms, p_nuc*Pa_2_MPa), fontsize=12)
    ax.set_ylim([0.001,1000])
    # plt.legend()

    # if save_figs:
    #     plt.savefig(save_folder + 'co2_bubble_psat_72bar_V360.pdf', bbox_inches='tight')

    # print out values of model at the measured points
    print('Model prediction at fit point is R = {0:f} um.'.format(np.interp(t_bubble, t_list[i], 
                                                                            np.array(R_list[i])*m_2_um)))
    print('Model prediction at addl point is R = {0:f} um.'.format(np.interp(t_bubble_2, t_list[i], 
                                                                             np.array(R_list[i])*m_2_um)))
    # prints the initial pressure (nucleation pressure) [MPa]
    print('Nucleation pressure is {0:.1f} MPa.'.format(p_list[i][0]*Pa_2_MPa))
```

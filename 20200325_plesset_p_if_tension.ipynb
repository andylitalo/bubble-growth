# Modeling Bubble Growth Down Microfluidic Flow-focusing Channel

When I can improve clarity and reduce motion blur sufficiently to measure the size of bubbles in the microfluidic device accurately, I will want to estimate when the bubbles nucleated as well to compare to nucleation barrier predictions from the Wang group. One way to estimate when a bubble nucleated is to estimate its growth leading up to the point of observation and using the point at which the bubble was the size of a critical nucleus (about 5 nm according to Huikuan's models) as the onset of bubble nucleation.

## Estimating Bubble Growth

Bubble growth is commonly modeled using the Epstein-Plesset equation (1950). One prominent example of its use was in the *Physics Today* article published by Zenit and Rodriguez-Rodriguez (2018) where the authors modeled the growth of bubbles in beer as they rose up in a glass of beer. This specific example permits several simplifying assumptions, such as neglecting the effect of surface tension since bubbles are large enough for the effect to be negligible (Epstein and Plesset found this to be the case for bubbles 100 um and larger in air-water solutions at room temperature) by the time they detach from their nucleation site on the walls, and assuming a constant pressure since the hydrostatic pressure in a glass of beer is negligible compared to atmospheric pressure. These assumptions, however, are not valid in the microfluidic flow-focusing channel used in our experiments.

In our experiments, bubbles nucleate at a nanoscopic size. At such small radii, the interfacial tension can produce Laplace pressures upwards of 40 bar. Additionally, the pressure changes dramatically (up to 100 bar) and on a similar time scale as the bubble growth due to the fast flow down the observation capillary. Therefore, we must incorporate the effects of interfacial tension and pressure changes into the bubble growth model.

## Assumptions

1. Diffusivity $D$ of $CO_2$ in the polyol and the interfacial tension $\gamma$ between the polyol-rich parent phase and $CO_2$-rich bubble are given by the equilibrium values at the current pressure in the microfluidic channel $p$
2. The pressure inside the bubble is given by the pressure in the parent phase $p$ plus the Laplacian pressure $p_L = \frac{2\gamma}{R}$
3. The density of $CO_2$ in the bubble is determined by the single-component $CO_2$ equation of state at the pressure inside the bubble (see #2) and the bubble radius instantly equilibrates to achieve this density
4. The concentration of $CO_2$ in the parent phase remains equal to the concentration under the initial saturation conditions due to the negligible size of bubbles relative to the parent phase during early stages of growth
5. The concentration of $CO_2$ at the interface of the bubble is the saturation pressure of $CO_2$ in polyol at equilibrium at the pressure in the parent phase
6. The pressure drop down the channel is governed by Poiseuille flow. We model the system as a sheath of a more viscous fluid surrounding a cylinder of a less viscous fluid. We assume that both fluids are Newtonian and incompressible.
7. The pressure at the outlet of the observation capillary is atmospheric pressure (this might not be exactly true due to the build-up of polyol at the outlet) 
8. Initial bubble size is given by the critical radius predicted by Dr. Huikuan Chao's string-method calculations
9. Constant temperature $T$
10. Fickian diffusion
11. Spherical symmetry


## Modeling Protocol

**NOTE: Initial versions of this notebook used `eps_pless_p_if_1`, which measured time starting at the entrance of the channel, which is appropriate for determining the pressure in the channel, but adapts the Epstein-Plesset model such that $\frac{\mathrm{d}m}{\mathrm{d}t} = 4\pi R^2 D \left(c_b - c_s(p)\right)\left(\frac{1}{R} + \frac{1}{\sqrt{\pi D (t+t_{nuc})}}\right)$, where $t_{nuc}$ is the time of nucleation and $t$ is measured from that point. This is not appropriate for the calculation of $\frac{\mathrm{d}m}{\mathrm{d}t}$ because the time used in this diffusive model should only be measured after nucleation. This error is corrected in `eps_pless_p_if_3`, which is now used throughout this notebook.**

The model iteratively solves the radius of the bubble and the mass of gas in the bubble after each time step. First, the amount of mass that diffuses into the bubble from the parent phase in a small time step is added to the total mass using the folllowing equation from equation (8) in Epstein and Plesset (1950)

\begin{align*}
    \frac{\mathrm{d}m}{\mathrm{d}t} &= 4\pi R^2 D \left(c_b - c_s(p)\right)\left(\frac{1}{R} + \frac{1}{\sqrt{\pi D t}}\right)\\
\end{align*}

where $m(t)$ is the mass of $CO_2$ inside the bubble at time $t$, $R(t)$ is the radius of the bubble, $D$ is diffusivity of $CO_2$ in the polyol, $c_s(p)$ is the saturation concentration of $CO_2$ in the polyol at pressure $p$, and $c_b$ is the concentration in the bulk of the parent phase. In our case, the bulk concentration is the saturation concentration at the pressure at which the system was saturated, $p_s$.

Next, we update the pressure based on the inlet pressure of the observation capillary calculated using Poiseuille flow,

\begin{align*}
    p &= p_{in} + \frac{vt}{L}(p_{atm} - p_{in})\\    
\end{align*}

where $p_{in}$ is the inlet pressure, $p_{atm}$ is the outlet pressure, which we have assumed to be atmospheric pressure (1 atm). We calculate the inlet pressure $p_{in}$, the velocity of the inner stream $v$, and the radius of the inner stream $R_i$ by solving the equations for Poiseuille flow modified to have a lower-viscosity inner stream (recorded in `20200323_cap_dim_calc.nb` Mathematica notebook and derived in my candidacy report, `YlitaloA_candidacy_report.pdf`),

\begin{equation}
    Q_o = \frac{\pi(p_{in}-p_{atm})(R_o^2 - R_i^2)^2}{8 \eta_o L}\\
\end{equation}
\begin{equation}
        \frac{p_{in} - p_{atm}}{L} = \frac{8 \eta_i Q_i}{\pi R_i^2\left[2(R_o^2 - R_i^2) \frac{\eta_i}{\eta_o} + R_i^2\right]}
\end{equation}
\begin{equation}
    v = \frac{p_{in} - p_{atm}}{L}\left( \frac{R_o^2 - R_i^2}{4\eta_o} + \frac{R_i^2}{4\eta_i} \right)\\
\end{equation}

where $\eta_o$ is the viscosity of the pure polyol in the outer stream, $\eta_i$ is the viscosity of the polyol-CO2 solution in the inner stream (usually crudely estimated due to lack of experimental measurements of the viscosity), $R_o$ is the radius of the outer stream (half the inner diameter of the observation capillary), $R_i$ is the radius of the inner stream, $Q_o$ is the flow rate of the outer stream, and $Q_i$ is the flow rate of the inner stream.

We then update the interfacial tension $\gamma(p)$ and the saturation concentration $c_s(p)$ according to experimental measurements. Finally, we self-consistently solve for the radius of the bubble under equilibrium conditions,

\begin{align*}
    R &= \left(\frac{3}{4\pi}\frac{m}{\rho_{CO2}(p + 2\gamma/R,T)}\right)^{1/3}\\
\end{align*}

where $\rho_{CO2}(p,T)$ gives the density of pure $CO_2$ at the pressure $p$ and temperature $T$. Note that we have added the Laplace pressure $2\gamma/R$ to the internal bubble pressure.

In this example, a CO2 bubble is artificially nucleated at the entrance of the observation capillary, where the pressure is about 80 bar, despite the saturation pressure being around 72 bar (below we identify a more reasonable nucleation pressure around 58 bar). The pressure inside the bubble is only noticeably higher than the pressure outside the bubble until the bubble starts to grow. This gives me confidence that the density inside the bubble will not be more vapor-like by the time it reaches 50 nm, which should allow us to detect it with SAXS. The initial nuclei, however, have liquid-like densities, and thus will be difficult to detect, especially given that they will occupy less total volume as well (so the intensity, which is proportional to the square of the density difference times the occupied volume of the bubbles, $I \propto (\Delta \rho V)^2$, will be lowered by both factors during the initial nucleation and growth).

The bubble growth skyrockets at the end as, fractionally, the pressure drop accelerates, as seen by the downturn of the pressure at the end. The final bubble is about 130 $\mu$m in radius, or 260 $\mu$m in diameter, roughly the size observed in polyurethane foams produced under similar conditions.


We now want to make sure that the results converge as the time step is decreased.

```python
dt_list = [0.0001, 0.0003, 0.001, 0.003, 0.01]
t_list = []
m_list = []
D_list = []
p_list = []
if_tension_list = []
c_s_list = []
R_list = []
rho_co2_list = []

# solves bubble growth for different time steps
for dt in dt_list:
    t, m, D, p, if_tension, c_s, R, rho_co2 = bubble.eps_pless_p_if_3(dt, t_nuc, p_s, R_nuc, p_atm, L, p_in, v,
                                                                  polyol_data_file, eos_co2_file)
    t_list += [t]
    m_list += [m]
    D_list += [D]
    p_list += [p]
    if_tension_list += [if_tension]
    c_s_list += [c_s]
    R_list += [R]
    rho_co2_list += [rho_co2]
    print('completed iteration for dt = {0:f}'.format(dt))
```

```python
# plots R(t)
fig = plt.figure()
ax = fig.add_subplot(111)
for i in range(len(dt_list)):
    dt = dt_list[i]
    t = t_list[i]
    R = R_list[i]
    ax.plot(np.array(t)*s_2_ms, np.array(R)*m_2_um, label='dt = {0:.1f} ms'.format(dt*s_2_ms))
    
ax.set_yscale('log')
ax.set_xlabel('time [ms]', fontsize=18)
ax.set_ylabel(r'$R(t)$ [$\mu$m]', fontsize=18)
ax.set_title(r'$R(t)$ of CO2 Bubble in V360 for Different Time Steps', fontsize=20)
plt.legend()
```

With a time step below 1 ms, the results converge, showing that the numerical scheme is stable and that our previous results (which used a time step of 1 ms) were reliable.

### Proof of Why the Diffusion Distance is Very Quickly Greater than the Depletion Thickness

On April 15, 2020, Prof. Julie Kornfield requested that I check that my model is not depleting $CO_2$ from the parent phase faster than it can diffuse into the bubble because she believes that the depletion of $CO_2$ around a bubble during growth has an important effect on the dynamics of bubble growth. I demonstrated that this was true in some plots above, but these only apply to the particular conditions that were considered in this notebook. Below, I prove that the depletion thickness is *always* smaller than the diffusion distance, and thus that the model is not violating diffusion limitations. To prove this, I prove an equivalent fact, which is that the mass that has diffused inside the bubble is smaller than the excess mass of $CO_2$ in the parent phase in a thickness of $\sqrt{Dt}$ outside the bubble, approximately the amount that could diffuse into the bubble in a diffusion-limited system.

#### Prove that:

\begin{align*}
    m < V_d \delta\\
\end{align*}

for all time $t > 0$, where $m$ is the mass of $CO_2$ inside the bubble, $V_d = \frac{4\pi}{3}\left[(R+\sqrt{Dt})^3 - R^3\right]$ is the volume that could be depleted by diffusion, $R$ is the radius of the bubble, $D$ is the diffusion constant of $CO_2$ in the parent phase, $\delta = c_b - c_s(p)$ is the excess mass-per-volume concentration of $CO_2$ in the parent phase above saturation, $c_b$ is the mass-per-volume concentration of $CO_2$ in the bulk of the parent phase (supersaturated, so $c_b > c_s(p)$), and $c_s(p)$ is the equilibrium saturation mass-per-volume concentration of $CO_2$ in the parent phase at a pressure of $p$.

To prove this statement, we prove that the time derivatives of each side follow this behavior, *i.e.*,

\begin{align*}
    \frac{\mathrm{d}m}{\mathrm{d}t} < \left(\frac{\mathrm{d}}{\mathrm{d}t}\right) V_d \delta \qquad \text(1) \\
\end{align*}

According to the Epstein-Plesset model cited at the beginning of this notebook, 

\begin{align*}
    \frac{\mathrm{d}m}{\mathrm{d}t} &= 4 \pi R^2 D \delta \left(\frac{1}{R} + \frac{1}{\sqrt{\pi D t}}\right)\\
\end{align*}

The time-derivative of the righthand side is

\begin{align*}
    \frac{\mathrm{d}(V_d \delta)}{\mathrm{d}t} &= \frac{\mathrm{d}V_d}{\mathrm{d}t}\delta + \frac{\mathrm{d}\delta}{\mathrm{d}t}V_d\\
\end{align*}

We first focus on the first term,

\begin{align*}
        \frac{\mathrm{d}V_d}{\mathrm{d}t}\delta &= \frac{4\pi \delta}{3}\left[3(R+\sqrt{Dt})^2 \left(\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{D}}{2 \sqrt{t}}\right) - 3R^2 \frac{\mathrm{d}R}{\mathrm{d}t}\right]\\
        &= 4\pi R^2 D \delta \left[\left(\frac{1}{\sqrt{D}}+\frac{\sqrt{t}}{R}\right)^2 \left(\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{D}}{2 \sqrt{t}}\right) - \frac{1}{D} \frac{\mathrm{d}R}{\mathrm{d}t}\right]\\
        &= 4\pi R^2 D \delta \left[\left(\frac{1}{R}+\frac{1}{\sqrt{Dt}}\right)^2 \left(t\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{Dt}}{2}\right) - \frac{1}{D} \frac{\mathrm{d}R}{\mathrm{d}t}\right]\\
        &= 4\pi R^2 D \delta \left\{\left[\left(\frac{1}{R}+\frac{1}{\sqrt{Dt}}\right)^2 t - \frac{1}{D}\right]\frac{\mathrm{d}R}{\mathrm{d}t} + \left(\frac{1}{R}+\frac{1}{\sqrt{Dt}}\right)^2\frac{\sqrt{Dt}}{2} \right\}\\
        &= 4\pi R^2 D \delta \left[\left(\frac{t}{R^2} + \frac{2 \sqrt{t}}{R \sqrt{D}}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{Dt}}{2R^2} + \frac{1}{R} + \frac{1}{2\sqrt{Dt}}\right]\\
        &= 4 \pi R^2 D \delta \left(\frac{1}{R} + \frac{1}{\sqrt{\pi D t}}\right) + 4\pi R^2 D \delta \left[\left(\frac{t}{R^2} + \frac{2 \sqrt{t}}{R \sqrt{D}}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{Dt}}{2R^2} + \frac{1}{2\sqrt{Dt}} - \frac{1}{\sqrt{\pi D t}}\right]\\
        &= \frac{\mathrm{d}m}{\mathrm{d}t} + 4\pi R^2 D \delta \left[\left(\frac{t}{R^2} + \frac{2 \sqrt{t}}{R \sqrt{D}}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{Dt}}{2R^2} + \frac{1}{2\sqrt{Dt}} - \frac{1}{\sqrt{\pi D t}}\right]\\
\end{align*}

We next simplify the second term,

\begin{align*}
    \frac{\mathrm{d}\delta}{\mathrm{d}t}V_d &= -\frac{\mathrm{d}c_s}{\mathrm{d}t}V_d\\
    &= V_d \left(\frac{\mathrm{d}c_s}{\mathrm{d}p}\right)\left(-\frac{\mathrm{d}p}{\mathrm{d}t}\right)\\
    &= \frac{4\pi}{3}\left[(R+\sqrt{Dt})^3 - R^3\right]\left(\frac{\mathrm{d}c_s}{\mathrm{d}p}\right)\left(-\frac{\mathrm{d}p}{\mathrm{d}t}\right)\\
\end{align*}

Combining the two terms, we have that the righthand side of equation (1) is

\begin{align*}
    \left(\frac{\mathrm{d}}{\mathrm{d}t}\right) V_d \delta &= m + 4\pi R^2 D \delta \left[\left(\frac{t}{R^2} + \frac{2 \sqrt{t}}{R \sqrt{D}}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{Dt}}{2R^2} + \frac{1}{2\sqrt{Dt}} - \frac{1}{\sqrt{\pi D t}}\right] + \frac{4\pi}{3}\left[(R+\sqrt{Dt})^3 - R^3\right]\left(\frac{\mathrm{d}c_s}{\mathrm{d}p}\right)\left(-\frac{\mathrm{d}p}{\mathrm{d}t}\right)\\
    &= \frac{\mathrm{d}m}{\mathrm{d}t} + 4\pi R^2 D \delta \left\{\left(\frac{t}{R^2} + \frac{2 \sqrt{t}}{R \sqrt{D}}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{Dt}}{2R^2} + \frac{1}{2\sqrt{Dt}} - \frac{1}{\sqrt{\pi D t}} + \frac{R}{3 D \delta}\left(\frac{\mathrm{d}c_s}{\mathrm{d}p}\right)\left(-\frac{\mathrm{d}p}{\mathrm{d}t}\right)\left[\left(1+\frac{\sqrt{Dt}}{R}\right)^3-1\right]\right\}\\
\end{align*}

Equation (1) is thus equivalent to

\begin{align*}
    \left(\frac{t}{R^2} + \frac{2 \sqrt{t}}{R \sqrt{D}}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{\sqrt{Dt}}{2R^2} + \frac{1}{2\sqrt{Dt}} - \frac{1}{\sqrt{\pi D t}} + \frac{R}{3 D \delta}\left(\frac{\mathrm{d}c_s}{\mathrm{d}p}\right)\left(-\frac{\mathrm{d}p}{\mathrm{d}t}\right)\left[\left(1+\frac{\sqrt{Dt}}{R}\right)^3-1\right] &> 0 \qquad (2)\\
\end{align*}

This expression is challenging to work with directly, so we will explore the behavior at the two extremes. First, we consider the case where $\frac{Dt}{R^2}\gg 1$. In this case, the one negative term in equation (2), $-\frac{1}{\sqrt{\pi D t}}$, is much smaller in magnitude than many of the other terms, such as $\frac{\sqrt{Dt}}{2R^2}$. Because the remaining terms are positive (note that $\frac{\mathrm{d}c_s}{\mathrm{d}p}>0$ because increasing pressure always increases saturation of $CO_2$ in polymer, and that $-\frac{\mathrm{d}p}{\mathrm{d}t}>0$ because the pressure decreases along the channel), the whole expression is positive, as desired.

We now turn to the other extreme, $\frac{Dt}{R^2}\ll 1$. In this case, equation (2) can be simplified roughly as

\begin{align*}
    \left(\frac{2 \sqrt{t}}{R \sqrt{D}}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \left(\frac{1}{2}-\frac{1}{\sqrt{\pi}}\right)\frac{1}{\sqrt{Dt}} + \frac{R}{3 D \delta}\left(\frac{\mathrm{d}c_s}{\mathrm{d}p}\right)\left(-\frac{\mathrm{d}p}{\mathrm{d}t}\right)\left(3\frac{\sqrt{Dt}}{R}\right) &> 0 \\
    \frac{1}{\sqrt{Dt}}\left[\left(\frac{2t}{R}\right)\frac{\mathrm{d}R}{\mathrm{d}t} + \left(\frac{1}{2}-\frac{1}{\sqrt{\pi}}\right) + \frac{t}{\delta}\frac{\mathrm{d}\delta}{\mathrm{d}t}\right] &> 0 \\
    \frac{1}{\sqrt{Dt}}\left[\left(\frac{2}{R}\frac{\mathrm{d}R}{\mathrm{d}t} + \frac{1}{\delta}\frac{\mathrm{d}\delta}{\mathrm{d}t}\right)t + \left(\frac{1}{2}-\frac{1}{\sqrt{\pi}}\right)\right] &> 0 \\
\end{align*}

The coefficient of the $t$ term is positive, so eventually this expression will be greater than 0 and will be true. However, at very early times, as $t \rightarrow 0^+$, it is possible that the expression on the lefthand side will not be greater than 0. Within this case, we consider two sub-cases: $\delta(t=0) = 0$ and $\delta(t=0) > 0$ ($\delta(t=0) < 0$ is not possible because a $\delta <0$ implies a sub-saturated solution, in which homogeneous bubble nucleation is not possible).

If $\delta(t=0) = 0$, then $\underset{t \rightarrow 0^+}{\lim}\left(\frac{\delta}{t}\right) = \frac{\mathrm{d}\delta}{\mathrm{d}t}$. Because $R(t=0) = R_{nuc} > 0$ and $\left.\frac{\mathrm{d}R}{\mathrm{d}t}\right|_{t=0} \propto \frac{1}{\sqrt{t}}$, the first term will go to 0 as $t \rightarrow 0^+$. Consequently, in the limit as $t \rightarrow 0^+$, the above inequality becomes

\begin{align*}
    \frac{1}{\sqrt{Dt}}\left[\left(0 + \frac{1}{\frac{\mathrm{d}\delta}{\mathrm{d}t}}\frac{\mathrm{d}\delta}{\mathrm{d}t}\right) + \left(\frac{1}{2}-\frac{1}{\sqrt{\pi}}\right)\right] &> 0 \\
    \frac{1}{\sqrt{Dt}}\left(\frac{3}{2}-\frac{1}{\sqrt{\pi}}\right) &> 0\\
\end{align*}

which is, indeed, true. This means that the mass accessible by diffusion will begin growing at a greater rate than the mass inside the bubble. This will continue as $t$ grows as described earlier, and will also be true when $\frac{Dt}{R^2} \gg 1$, and so we expect the mass of $CO_2$ accessible by diffusion always to be more than the mass inside the bubble (minus the initial mass).

If, on the other hand $\delta(t=0) > 0$, then in the limit as $t \rightarrow 0^+$, the inequality becomes

\begin{align*}
    \frac{1}{\sqrt{Dt}}\left[0 + \left(\frac{1}{2}-\frac{1}{\sqrt{\pi}}\right)\right] &> 0 \\
    \frac{1}{\sqrt{Dt}}\left(\frac{1}{2}-\frac{1}{\sqrt{\pi}}\right) &> 0\\
\end{align*}

which is not true. **Therefore, we expect that when the bubble nucleates in a supersaturated solution (** *i.e.,* $\delta(t =0) > 0$**), the rate of increase of mass in the bubble may briefly exceed the rate at which $CO_2$ is expected to become available through diffusion.** This was seen in the plots above where the diffusion time was larger than the time since nucleation for the first few tenth of a millisecond roughly. Nevertheless, as mentioned before, this inequality will be made true at some point in time, which appears to be very short compared to the overall lifetime of the bubble.
